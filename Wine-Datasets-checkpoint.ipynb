```python

plt.figure(figsize=(15,5))
df1=sns.barplot(x="quality",y="citric acid",data=df)



for bars in df1.containers:
    df1.bar_label(bars)
    
plt.title('citric acid vs quality')


plt.show()
```

```python

df2=sns.displot(df["quality"])
plt.show()
```

```python
df3=sns.barplot(x="quality",
               y="volatile acidity",data=df,hue="quality",
                palette= "CMRmap",alpha=1)

plt.title("volatile vs quality")
plt.show()
```

```python
df4=sns.barplot(x="quality",
               y="chlorides",data=df,hue="quality",
                palette= "CMRmap",alpha=1)

plt.title("chlorides vs quality")
plt.show()
```

```python
df5=sns.barplot(x="quality",
               y="sulphates",data=df,hue="quality",
                palette= "CMRmap",alpha=1)

plt.title("sulphates vs quality")
plt.show()
```

```python
plt.figure(figsize=(15,5))
df6=sns.factorplot(x="alcohol",y="density",data=df_1,kind="box")
plt.grid()
plt.show()
```

```python
df7=sns.catplot(x="quality",data=df,palette="BuPu",kind="count") #box,point,etc
plt.grid()
plt.show()
```

```python
correlation=df.corr()
plt.figure(figsize=(15,15))
y={"fontsize":5,"color":"g"}
df8=sns.heatmap(correlation,cbar=True, annot=True,cmap="PuBuGn", annot_kws=y,
                linecolor="y",yticklabels=True,xticklabels=True,fmt="2")
plt.show()
```

The higher the quality the better the alcohol.
The higher the volatile acidity ,the leser the quality of alcohol.

The higher the clorides the lesser the quality.
The higher the sulphates the better the quality.
